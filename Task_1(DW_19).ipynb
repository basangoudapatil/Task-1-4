```python
#File Handling Assignment
print('1:Sign Up \n 2: Sign In \n 3: Forgot Password')
n=int(input('Enter your Input:'))

if n==1:#Sign Up

  import re

  username=input('Enter your Username: ')
  result=re.compile('[a-zA-Z0-9]+@+[a-zA-Z]+\.+[a-zA-Z]+')
  answer=result.findall(username)

  flag=True
  if (answer):
    print('Username is Valid')
    flag=True
  else:
    print('Username is not Valid')
    flag=False
  print(answer)

  password=input('Enter your Password: ')
  result1=re.compile('[a-zA-Z0-9]+[!@#$%^&*()]+')
  answer1=result1.findall(password)

  flag=True
  if (answer1):
    print('Password is Valid')
    flag=True
  else:
    print('Password is not Valid')
    flag=False
  print(answer1)

  if flag==True:
    F1=open('Username1','a')
    F1.write('\n')
    F1.writelines(answer)
    F1.close()
    F2=open('Password1','a')
    F2.write('\n')
    F2.writelines(answer1)
    F2.close()

if n==2: #Sign In

  flag=True
  username=input('Enter your Username: ')
  F1=open('Username1','r')
  for name in F1:
    if name==username:
      print('Valid User')
      flag=True
      break
  else:
    print('Invalid User')
    flag=False
      
  if flag==True:
    password=input('Enter your Password: ')
    F2=open('/content/Password1','r')
    for pas in F2:
      if pas==password:
        print('Logged In')
        break
    else:
        print('Invalid Password!! \n Try again ')
    F2.close()
  F1.close()

elif n==3: #forgot password
  
  import re
  
  username=input('Enter your Username: ')
  password=input('Enter your previous password: ')
  flag=True
  F1=open('/content/Username1','r')
  F2=open('/content/Password1','r+')
  for name in F1:
    if name==username:
      flag=True
      break
  else:
    flag=False

  if flag==True:
    flag=True
    for pas in F2:
      if pas==password:
        print('Password is valid')
        flag=True
        break
    else:
      print('Password is not valid \n Enter new password: ')
      flag=False

  if flag==False:
    password=input('Enter new password')
    result1=re.compile('[a-zA-Z0-9]+[!@#$%^&*()]+')
    answer1=result1.findall(password)

  if (answer1):
    print('Password is Valid')
    F2.write('\n')
    F2.writelines(answer1)
    F2.close()
  else:
    print('Password is not Valid')
  print(answer1)
  F1.close()
```
